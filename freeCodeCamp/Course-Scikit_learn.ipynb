# Sklearn - KNN example

```python
import pandas as pd
import numpy as np

df = pd.read_csv('car.data', names=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class'])
print(df.shape)
print(df.info())
df.head()
```

```python
X = df[['buying', 'maint', 'safety']]
y = df[['class']]
print(X.shape)
print(y.shape)
```

```python
from sklearn import neighbors, metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# X preprocessing
X_values = X.values
Le = LabelEncoder()
for i in range(len(X_values[0])):
  X_values[:, i] = Le.fit_transform(X_values[:, i])
  print('Class {}: '.format(X.columns[i]), Le.classes_)
print('\nData: \n', X_values[:2])
```

```python
label_mapping = {
    'unacc': 0,
    'acc': 1,
    'good': 2,
    'vgood': 3
}
y_values = y['class'].map(label_mapping)
print(type(y_values))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_values, y_values, test_size=0.2)
```

```python
# create model
knn = neighbors.KNeighborsClassifier(n_neighbors=5, weights='uniform')
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

acc = metrics.accuracy_score(pred, y_test)
acc
```

```python
import pandas as pd
import mnist
from PIL import Image
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
```

```python
# training
X_train = mnist.train_images()
y_train = mnist.train_labels()

# test
X_test = mnist.test_images()
y_test = mnist.test_labels()
```

```python
print(X_train.shape)
print(X_test.shape)

X_train = X_train.reshape((-1, 28*28))
X_test = X_test.reshape((-1, 28*28))

print(X_train.shape)
print(X_test.shape)
```

```python
print(X_train[0])
```

```python
# to minimize CPU need, let's divide our pixel by 256
X_train = X_train/256
X_test = X_test/256

print(X_train[0])
```

```python
clf = MLPClassifier(solver='adam', activation='relu', hidden_layer_sizes=(64,64))
clf.fit(X_train, y_train)
```

```python
pred = clf.predict(X_test)
```

```python
acc = confusion_matrix(y_test, pred)
acc
```

```python
def accuracy(cm):
  diagonal = cm.trace()
  elements = cm.sum()
  return diagonal/elements
```

```python
accuracy(acc)
```
