```python
resp_lamp = requests.get('https://data.covid19.go.id/public/api/prov_detail_LAMPUNG.json')
cov_lamp_raw = resp_lamp.json()
cov_lamp_raw
```

```python
print('Nama-nama elemen utama:\n', cov_lamp_raw.keys())
print('\nJumlah total kasus COVID-19 di Lampung                 : %d' %cov_lamp_raw['kasus_total'])
print('Persentase kematian akibat COVID-19 di Lampung         : %f.2%%' %cov_lamp_raw['meninggal_persen'])
print('Persentase tingkat kesembuhan dari COVID-19 di Lampung : %f.2%%' %cov_lamp_raw['sembuh_persen'])
```

```python
import numpy as np
import pandas as pd
cov_lamp = pd.DataFrame(cov_lamp_raw['list_perkembangan'])
print('Info cov_lamp:\n', cov_lamp.info())
print('\nLima data teratas cov_lamp:\n')
cov_lamp.head()
```

```python
cov_lamp_tidy = (cov_lamp.drop(columns=[item for item in cov_lamp.columns 
                                               if item.startswith('AKUMULASI') 
                                                  or item.startswith('DIRAWAT')])
                           .rename(columns=str.lower)
                           .rename(columns={'kasus': 'kasus_baru'})
                  )
cov_lamp_tidy['tanggal'] = pd.to_datetime(cov_lamp_tidy['tanggal']*1e6, unit='ns')
print('Lima data teratas:\n', cov_lamp_tidy.head())
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.clf()
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(data=cov_lamp_tidy, x='tanggal', height='kasus_baru')
plt.show()
```

```python
plt.clf()
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(data=cov_lamp_tidy, x='tanggal', height='kasus_baru', color='salmon')
fig.suptitle('Kasus Harian Positif COVID-19 di Lampung', 
             y=1.00, fontsize=16, fontweight='bold', ha='center')
ax.set_title('Terjadi pelonjakan kasus di awal bulan Juli akibat klaster Secapa AD Bandung',
             fontsize=10)
ax.set_xlabel('')
ax.set_ylabel('Jumlah kasus')
ax.text(1, -0.3, 'Sumber data: covid.19.go.id', color='blue',
        ha='right', transform=ax.transAxes)
ax.set_xticklabels(ax.get_xticks(), rotation=90)

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
cov_lamp_pekanan = (cov_lamp_tidy.set_index('tanggal')['kasus_baru']
                                   .resample('W')
                                   .sum()
                                   .reset_index()
                                   .rename(columns={'kasus_baru': 'jumlah'})
                    )
cov_lamp_pekanan['tahun'] = cov_lamp_pekanan['tanggal'].apply(lambda x: x.year)
cov_lamp_pekanan['pekan_ke'] =cov_lamp_pekanan['tanggal'].apply(lambda x: x.weekofyear)
cov_lamp_pekanan = cov_lamp_pekanan[['tahun', 'pekan_ke', 'jumlah']]

print('Info cov_lamp_pekanan:')
print(cov_lamp_pekanan.info())
print('\nLima data teratas cov_lamp_pekanan:\n', cov_lamp_pekanan.head())
```

```python
cov_lamp_pekanan['jumlah_pekanlalu'] = cov_lamp_pekanan['jumlah'].shift().replace(np.nan, 0).astype(np.int)
cov_lamp_pekanan['lebih_baik'] = cov_lamp_pekanan['jumlah'] < cov_lamp_pekanan['jumlah_pekanlalu']

print('Sepuluh data teratas:\n', cov_lamp_pekanan.head(10))
```

```python
plt.clf()
jml_tahun_terjadi_covid19 = cov_lamp_pekanan['tahun'].nunique()
tahun_terjadi_covid19 = cov_lamp_pekanan['tahun'].unique()
fig, axes = plt.subplots(nrows=jml_tahun_terjadi_covid19, figsize=(10,3*jml_tahun_terjadi_covid19))

fig.suptitle('Kasus Pekanan Positif COVID-19 di Lampung',
y=1.00, fontsize=16, fontweight='bold', ha='center')
for i, ax in enumerate(axes):
  ax.bar( data=cov_lamp_pekanan.loc[cov_lamp_pekanan['tahun'] == tahun_terjadi_covid19[i]], 
        x='pekan_ke', height='jumlah', 
        color=['mediumseagreen' if x is True else 'salmon' for x in cov_lamp_pekanan['lebih_baik']])
  
  if i == 0:
    ax.set_title('Kolom hijau menunjukan penambahan kasus baru lebih sedikit dibandingkan satu pekan sebelumnya', fontsize=10)
  elif i == jml_tahun_terjadi_covid19-1:
    ax.text(1, -0.2, 'Sumber data: covid.19.go.id', color='blue', ha='right', transform=ax.transAxes)
  
  ax.set_xlim([0, 52.5])
  ax.set_ylim([0, max(cov_lamp_pekanan['jumlah'])])
  ax.set_xlabel('')
  ax.set_ylabel('Jumlah kasus %d' %(tahun_terjadi_covid19[i],))
  ax.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
cov_lamp_akumulasi = cov_lamp_tidy[['tanggal']].copy()
cov_lamp_akumulasi['akumulasi_aktif'] = (cov_lamp_tidy['kasus_baru'] - cov_lamp_tidy['sembuh'] - cov_lamp_tidy['meninggal']).cumsum()
cov_lamp_akumulasi['akumulasi_sembuh'] = cov_lamp_tidy['sembuh'].cumsum()
cov_lamp_akumulasi['akumulasi_meninggal'] = cov_lamp_tidy['meninggal'].cumsum()
print(cov_lamp_akumulasi.tail())
```

```python
plt.clf()
fig, ax = plt.subplots(figsize=(10,5))
ax.plot('tanggal', 'akumulasi_aktif', data=cov_lamp_akumulasi, lw=2)

ax.set_title('Akumulasi aktif COVID-19 di Lampung',
             fontsize=22)
ax.set_xlabel('')
ax.set_ylabel('Akumulasi aktif')
ax.text(1, -0.3, 'Sumber data: covid.19.go.id', color='blue',
        ha='right', transform=ax.transAxes)
ax.set_xticklabels(ax.get_xticks(), rotation=90)

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.clf()
fig, ax = plt.subplots(figsize=(10, 5))
cov_lamp_akumulasi_ts = cov_lamp_akumulasi.set_index('tanggal')
cov_lamp_akumulasi_ts.plot(kind='line', ax=ax, lw=3, color=['salmon', 'slategrey', 'olivedrab'])
ax.set_title('Dinamika Kasus COVID-19 di Lampung', fontsize=22)
ax.set_xlabel('')
ax.set_ylabel('Akumulasi aktif')
ax.text(1, -0.3, 'Sumber data: covid.19.go.id', color='blue', ha='right', transform=ax.transAxes)

plt.grid()
plt.tight_layout()
plt.show()
```
