```python
tweets = tweepy.Cursor(api.search, q='#climate-change -filter:retweets', lang='en', since='2018-11-01').items(5)
all_tweets = [tweet.text for tweet in tweets]
all_tweets[:5]
```

```python
import re
def remove_url(txt):
    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())
all_tweets_no_urls = [remove_url(tweet) for tweet in all_tweets]
all_tweets_no_urls[:5]
```

```python
words_in_tweet = [tweet.lower().split() for tweet in all_tweets_no_urls]
words_in_tweet[:1]
```

```python
import itertools
import collections

all_words_no_urls = list(itertools.chain(*words_in_tweet))
counts_no_urls = collections.Counter(all_words_no_urls)
counts_no_urls.most_common(5)
```

```python
clean_tweets = pd.DataFrame(counts_no_urls.most_common(10),
                           columns=['words', 'count'])
clean_tweets.head()
```

```python
import matplotlib.pyplot as plt
clean_tweets.sort_values('count').plot.barh(x='words', y='count')
plt.show()
```
